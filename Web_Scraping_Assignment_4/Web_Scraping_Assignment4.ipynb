Q.1 - Scrape the details of most viewed videos on YouTube from Wikipedia.Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: (A) Rank (B) Name (C) Artist (D) Upload date (E) Views

```python
# Activating the chrome browser
driver = webdriver.Chrome('chromedriver.exe')
time.sleep(2)

# Opening the wikipedia
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)

time.sleep(2)

# Scraping rank of the video
rank = []
try:
    ranks = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
    for i in ranks:
        rank.append(i.text)
except NoSuchElementException:#handling no such element exception
    rank.append('No details available')
except StaleElementReferenceException:#handling Stale element exception
    rank.append('No details available')
    
time.sleep(2)

# Scraping name of the video
name = []
try:
    names = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
    for i in names:
        name.append(i.text)
except NoSuchElementException:#handling no such element exception
    name.append('No details available')
except StaleElementReferenceException:#handling Stale element exception
    name.append('No details available')
    
time.sleep(2)

# Scraping artist of the video
artist = []
try:
    artists = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
    for i in artists:
        artist.append(i.text)
except NoSuchElementException:#handling no such element exception
    artist.append('No details available')
except StaleElementReferenceException:#handling Stale element exception
    artist.append('No details available')

time.sleep(2)

# Scraping view of the video
view = []
try:
    views = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
    for i in artists:
        view.append(i.text)
except NoSuchElementException:#handling no such element exception
    view.append('No details available')
except StaleElementReferenceException:#handling Stale element exception
    view.append('No details available')

time.sleep(2)

# Scraping upload date of the video
upload_date = []
try:
    upload_dates = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
    for i in upload_dates:
        upload_date.append(i.text)
except NoSuchElementException:#handling no such element exception
    upload_date.append('No details available')
except StaleElementReferenceException:#handling Stale element exception
    upload_date.append('No details available')

time.sleep(2)

# Finally create a dataframe of the scraped data.
df=pd.DataFrame({"Rank":rank,
                 "Name":name,
                 "artist":artist,
                 "Views(in billion)":views,
                 "Date of upload":upload_date
                })
df
```

st ODI)

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 

time.sleep(2)

# Opening the bcci.tv
url = "https://www.bcci.tv/"
driver.get(url)

time.sleep(2)
```

```python
#going to international fixtures page
fixtures=driver.find_element_by_xpath("//div[@class='collapse navbar-collapse']/ul[1]/li[2]/a")
try:
    fixtures.click()
except ElementNotInteractableException:
    driver.get(fixtures.get_attribute('href'))
    
time.sleep(2)
```

```python
# make empty lists
match=[] 
series=[]
place=[]
date=[]
Time=[]
```

```python
#scraping time
time_title=driver.find_elements_by_xpath("//h5[@class='text-right ng-binding']")
for i in time_title:
    if i.text is None :
        Time.append("--") 
    else:
        Time.append(i.text)
```

```python
# scrape match details urls from xpath    
urls=[]
match_page=driver.find_elements_by_xpath("//a[@class='match-center-btn ng-scope']")  
for i in match_page:
    urls.append(i.get_attribute('href'))

time.sleep(2)

for i in urls:                 #list of urls of match description pages
    driver.get(i)
    
    time.sleep(2)
    
    #scarping match title
    try:                    
        match_title=driver.find_element_by_xpath("//div[@class='widget matchDetails hideMobOnly showForSpecialCase']/div/ul/li[2]/span[2]")
        match.append(match_title.text)
    except NoSuchElementException:#handling no such element exception
        match.append('No details available')
    
    time.sleep(2)
    #scarping series title    
    try:
        series_title=driver.find_element_by_xpath("//div[@class='widget matchDetails hideMobOnly showForSpecialCase']/div/ul/li[1]/span[2]")
        series.append(series_title.text)
    except NoSuchElementException:#handling no such element exception
        series.append("No details Available")
    
    time.sleep(2)
    
    #scarping date of match 
    try:
        date_title=driver.find_element_by_xpath("//div[@class='matchDate alignC ng-binding ng-scope']")
        date.append(date_title.text)
    except NoSuchElementException:#handling no such element exception
        date.append("No details Available")      
    
    time.sleep(2)
    
     #scarping match venue    
    try:
        place_title=driver.find_element_by_xpath("//div[@class='widget matchDetails hideMobOnly showForSpecialCase']/div/ul/li[3]/span[2]")
        place.append(place_title.text)
    except NoSuchElementException:#handling no such element exception
        place.append("No details Available")
        

#preparing the dataframe
df=pd.DataFrame({"Match":match,
                "Series":series,
                "Place":place,
                "Date":date,
                "Time":Time})
df  
```

Q5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used
Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(2)

# Opening the github
url = "https://github.com/"
driver.get(url)
```

```python
# clicking on explore option button
explore=driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary")
explore.click()
time.sleep(2)

#then finding trending option and click
trending=driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a")
try:
    trending.click()
except:
    driver.get(trending.get_attribute('href'))
time.sleep(2)
```

```python
# making empty lists    
repo_urls=[]
repo_title=[]
repo_desc=[]
contributors_count=[]
language_used=[]

#scraping repositories urls
repos=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for i in repos:
    repo_urls.append(i.get_attribute('href'))
time.sleep(2)

#scraping repositories title
repo_title=[]
try:
    repos=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']")
    for i in repos:
        repo_title.append(i.text)
except NoSuchElementException:#handling no such element exception
    repo_title.append('No details available')
time.sleep(3)

for i in repo_urls:
    driver.get(i)
    time.sleep(3)
    
    #scraping repositories discription
    try:          
        repo_desc.append(driver.find_element_by_xpath("//p[@class='f4 my-3']").text)
    except:
        repo_desc.append('-')
        
    #scraping contributors count
    try:
        count=driver.find_element_by_xpath("/html/body/div[4]/div/main/div[2]/div/div/div[3]/div[2]/div/div[6]/div/h2/a/span")
        contributors_count.append(count.text)
    except NoSuchElementException:#handling no such element exception
        contributors_count.append('No details available')
    time.sleep(2)
    
    #scraping languages used
    language_list=[]
    languages=driver.find_elements_by_xpath("//li[@class='d-inline']/a/span[1]")
    if languages:    
        for i in languages:
            language_list.append(i.text)
    else:
        language_list.append('No languages used')
    language_used.append(language_list)

time.sleep(2)   
#creating dataframe
df=pd.DataFrame({'Title':repo_title,
                'Description':repo_desc,
                'Contributors_count':contributors_count,
                'Language_used':language_used})
df
```

```python
#scraping dataset names
name=[]
try:
    names=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
    for i in names:
        name.append(i.text)
except NoSuchElementException:#handling no such element exception
    name.append('No details available')
time.sleep(2)

#scraping dataset datatype
data_type=[]
try:
    datatypes=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]")
    for i in datatypes[1:]:
         data_type.append(i.text)
except NoSuchElementException:
    data_type.append("No information")
time.sleep(2)


#scraping dataset default task
default_task=[]
try:
    info=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]")
    for i in info[1:]:
         default_task.append(i.text)
except NoSuchElementException:
    default_task.append("No information")
time.sleep(2)


#scraping dataset attribute type
attr_type=[]
try:
    attr_types=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]")
    for i in attr_types[1:]:
         attr_type.append(i.text)
except NoSuchElementException:
    attr_type.append("No information")
time.sleep(2)


#scraping dataset no. of instance
no_of_instances=[]
try:
    info=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]")
    for i in info[1:]:
         no_of_instances.append(i.text)
except NoSuchElementException:
    no_of_instances.append("No information")
time.sleep(2)


#scraping dataset no. of attributes
no_of_attributes=[]
try:
    info=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]")
    for i in info[1:]:
         no_of_attributes.append(i.text)
except NoSuchElementException:
    no_of_attributes.append("No information")
time.sleep(2)


#scraping dataset year
year=[]
try:
    years=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]")
    for i in years[1:]:
         year.append(i.text)
except NoSuchElementException:
    year.append("No information")
time.sleep(2)

#creating dataframe
df=pd.DataFrame({'Name':name,
                'Data_type':data_type,
                'Task':default_task,
                'Attribute_type':attr_type,
                'No_of_instances':no_of_instances,
                'No_of_attributes':no_of_attributes,
                'Year':year})
df
```
