# Data Preprocessing

# Training process begins

### Finding best random_state

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
```

```python
maxAccu = 0
maxRS = 0

for i in range(1,200):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .20, random_state = i)
    rfc = RandomForestClassifier()
    rfc.fit(x_train,y_train)
    pred = rfc.predict(x_test)
    acc = accuracy_score(y_test, pred)
    print('accuracy',acc,'random_state',i)
    
    if acc > maxAccu:
        maxAccu = acc
        maxRS = i
```

```python
print("Best accuracy is ",maxAccu,"On Random_state ",maxRS)
```

```python
maxAccu = 0
maxRS = 0

for i in range(1,200):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .20, random_state = i)
    lr = LogisticRegression()
    lr.fit(x_train,y_train)
    pred = lr.predict(x_test)
    acc = accuracy_score(y_test, pred)
    print('accuracy',acc,'random_state',i)
    
    if acc > maxAccu:
        maxAccu = acc
        maxRS = i
```

```python
print("Best accuracy is ",maxAccu,"On Random_state ",maxRS)
```

```python
maxAccu = 0
maxRS = 0

for i in range(1,200):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .20, random_state = i)
    dtc = DecisionTreeClassifier()
    dtc.fit(x_train,y_train)
    pred = dtc.predict(x_test)
    acc = accuracy_score(y_test, pred)
    print('accuracy',acc,'random_state',i)
    
    if acc > maxAccu:
        maxAccu = acc
        maxRS = i
```

```python
print("Best accuracy is ",maxAccu,"On Random_state ",maxRS)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
pred = rfc.predict(x_test)
print('accuracy',accuracy_score(y_test, pred)*100)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)
pred = dtc.predict(x_test)
print('accuracy',accuracy_score(y_test, pred)*100)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

# SVC

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(x_train,y_train)
pred = svc.predict(x_test)
print('accuracy',accuracy_score(y_test, pred)*100)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

# Cross Validation

```python
from sklearn.model_selection import cross_val_score

scr = cross_val_score(rfc, x, y, cv=5)
print('cross validation score of random forest classifier model :',scr.mean())
```

```python
scr2 = cross_val_score(dtc, x, y, cv=5)
print('cross validation score of decision tree classifier model :',scr.mean())
```

```python
scr3 = cross_val_score(svc, x, y, cv=5)
print('cross validation score of support vector classifier model :',scr3.mean())
```

# Hyper parameter tuning

```python
from sklearn.model_selection import GridSearchCV

parameters = {'max_features': ['auto','sqrt','log2'],
              'max_depth': [4,5,6,7,8],
              'criterion':['gini','entropy']}
```

```python
GCV = GridSearchCV(RandomForestClassifier(),parameters,cv=5,scoring='accuracy')
GCV.fit(x_train,y_train)
GCV.best_params_
```
