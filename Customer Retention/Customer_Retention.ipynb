```python
#create new dataframe for pie chart.
longer_delivery_period = df['Longer delivery period'].value_counts().rename_axis('sub_delivery_period').reset_index(name='counts')
longer_delivery_period
```

```python
#lables and counts of longer delivery period variable.
your_labels = longer_delivery_period.sub_delivery_period
your_values = longer_delivery_period.counts
print(your_labels)
print(your_values)
```

```python
#display pie chart of target variable.
fig = plt.figure(figsize=(15,5))
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(your_values,labels=your_labels,autopct='%2.2f%%')
plt.legend()
plt.show()
```
