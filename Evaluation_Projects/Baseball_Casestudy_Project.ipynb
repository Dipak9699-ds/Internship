# Data Preprocessing

```python
#Splitting the independent and target variable in x and y
x = df.drop("W",axis=1)
y = df['W']
```

```python
x.head()
```

```python
print(y.head())
```

# Scaling the data

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()

from sklearn.linear_model import LinearRegression
lr = LinearRegression()

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()

from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()

from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
for i in range(1,100):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .20, random_state = i)
    lr.fit(x_train,y_train)
    pred_train = lr.predict(x_train)
    pred_test = lr.predict(x_test)
    
    train_acc = print(f"At random state {i}, the trainning accuracy is:- {r2_score(y_train,pred_train)}")
    test_acc = print(f"At random state {i}, the testing accuracy is:- {r2_score(y_test,pred_test)}")
    print("\n")
```

```python
for i in range(1,100):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .20, random_state = i)
    rf.fit(x_train,y_train)
    pred_train = rf.predict(x_train)
    pred_test = rf.predict(x_test)
    
    train_acc = print(f"At random state {i}, the trainning accuracy is:- {r2_score(y_train,pred_train)}")
    test_acc = print(f"At random state {i}, the testing accuracy is:- {r2_score(y_test,pred_test)}")
    print("\n")
```

```python
for i in range(1,100):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .20, random_state = i)
    knn.fit(x_train,y_train)
    pred_train = knn.predict(x_train)
    pred_test = knn.predict(x_test)
    
    train_acc = print(f"At random state {i}, the trainning accuracy is:- {r2_score(y_train,pred_train)}")
    test_acc = print(f"At random state {i}, the testing accuracy is:- {r2_score(y_test,pred_test)}")
    print("\n")
```

# Creating train-test split

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.20,random_state = 30)
```

```python
lr.fit(x_train,y_train)
```

```python
pred_test = lr.predict(x_test)
pred_test
```

```python
print(r2_score(y_test,pred_test))
```

```python
#predict the value
print("predicted result price:",pred_test)
print("actual price",y_test)
```

```python
#display MAE, MSE and RMSE 
from sklearn.metrics import mean_squared_error, mean_absolute_error
print('error:')

print('mean absolute error',mean_absolute_error(y_test,pred_test))
print('mean squared error',mean_squared_error(y_test,pred_test))

print('Root mean squared error',np.sqrt(mean_squared_error(y_test,pred_test)))
```

```python
train_accuracy = r2_score(y_train,pred_train)
test_accuracy = r2_score(y_test,pred_test)

from sklearn.model_selection import cross_val_score
for j in range(2,10):
    cv_score = cross_val_score(lr,x,y,cv=j)
    cv_mean = cv_score.mean()
    print(f"At cross fold {j} the cv score is {cv_mean} and accuracy score for trainning is {train_accuracy} and accuracy score for testing is {test_accuracy}")
    print("\n")
```
